The topic of classifier in today's lab, SVMs make for really good linear separators. Let us look at an example which has linearly separable data:

```python
np.random.seed(0)

ar = np.vstack(     [\
                    np.random.multivariate_normal(np.array([1, 1]), 1.5 * np.array([[2, -1], [-1, 2.0]]), size = 50, ),\
                    np.random.multivariate_normal(np.array([3, 3]), 2 * np.array([[0.75, -0.5], [-0.5, 0.75]]), size = 50, )
                    ]\
              )

testAr = np.vstack(   [\
                      np.random.multivariate_normal(np.array([1, 1]), np.array([[0.5, -0.25], [-0.25, 0.5]]), size = 500, ),\
                      np.random.multivariate_normal(np.array([3, 3]), np.array([[0.75, -0.5], [-0.5, 0.75]]), size = 500, )
                      ]\
                  )
testy = np.array([0] * int((testAr.shape[0]/2)) + [1] * int((testAr.shape[0]/2)))

X = ar
y = np.array([0] * int((ar.shape[0]/2)) + [1] * int((ar.shape[0]/2)))

def plotDecisionBoundary(X, y, pair, clf):
  x_min, x_max = X[:, pair[0]].min() - 1, X[:, pair[0]].max() + 1
  y_min, y_max = X[:, pair[1]].min() - 1, X[:, pair[1]].max() + 1
  xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                      np.arange(y_min, y_max, 0.1))

  y_pred = clf.predict(np.c_[xx.ravel(), yy.ravel()])
  y_pred = y_pred.reshape(xx.shape)
  plt.figure(figsize=(5,4))
  plt.contourf(xx, yy, y_pred, alpha=0.4)
  plt.scatter(X[:, pair[0]], X[:, pair[1]], c = y, s = 50, edgecolor='k')
  # plt.legend()
  plt.gcf().set_dpi(130)
  plt.show()

def boundaryExp() :
  clf = svm.LinearSVC()
  pair = [0, 1]
  clf.fit(X[:, pair], y)
  plotDecisionBoundary(X, y, pair, clf)

  plt.show()

boundaryExp()
```

## Intuitive introduction to SVMs:

![SVM](http://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1526288453/index3_souoaz.png)

SVMs are an iterative approach of trying to find the optimal hyperplane that divides the multidimentional space into different classes present in the dataset.


**Hyperplanes:** These are the decision planes that separate the objects of classes that we are trying to classify.

**Support Vectors** : Support vectors are the points from dataset that are closest to the hyperplane that divides the dataset.

**Margin**:  The gap between the closest support vectors from the different class along the direction perpendicular to the hyperplane. Simply put, it is the sum of perpendicular distance of the support vector of each class to the hyperplane.

# The Kernel Trick

The true potential of SVMs is unleashed when they are combined with kernels.

## Kernels : An intuitive explanation

Kernel methods are essentially counting on using the training data (say $i^{th}$ example $(x_i, y_i)$ ) itself in a more straightforward way and learning a corresponding weight ($w_i$) for that example. Rather than trying to learn a fixed set of parameters which is done typically.
Depending on the kind of kernel used, we can virtually project the training data in a higher dimension to make it easier for the classifier to classify them.


```python
from sklearn.datasets import make_circles
X, y = make_circles(200, factor=.1, noise=.1)

clf = svm.SVC(kernel='linear').fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='tab20')
plt.gcf().set_dpi(110)
```

```python
plt.style.use("default")
clf = svm.LinearSVC()
pair = [0, 1]
clf.fit(X[:, pair], y)
plotDecisionBoundary(X, y, [0, 1], clf)
```

However, if we artificially add another dimention to the dataset of the form:

$z = x^2 + y^2$

```python
Z = np.array([[i[0]**2 + i[1]**2] for i in X])
```

```python
X_new = np.hstack((X, Z))
```

That simple trick has helped us to get another dimension in which the data is linearly separable by a hyperplane (in this case, a 2d plane)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Define feature pairs and kernels
feature_pairs = [(0, 1), (0, 2), (0, 3), (1, 2), (1, 3), (2, 3)]
kernels = ['linear', 'poly', 'rbf', 'sigmoid', 'precomputed']

# Function to plot decision boundary
def plot_decision_boundary(X, y, pair, clf, kernel):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

    if kernel == 'precomputed':
        # For precomputed kernel, we must use X as it is (not the grid points)
        return

    y_pred = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    y_pred = y_pred.reshape(xx.shape)

    plt.figure(figsize=(6, 5))
    plt.contourf(xx, yy, y_pred, alpha=0.3, cmap='coolwarm')
    plt.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolor='k', cmap='coolwarm')
    plt.title(f'SVM Decision Boundary (Kernel: {kernel}) for Features: {pair}')
    plt.xlabel(f'Feature {pair[0]} ({iris.feature_names[pair[0]]})')
    plt.ylabel(f'Feature {pair[1]} ({iris.feature_names[pair[1]]})')
    plt.gcf().set_dpi(130)
    plt.show()

# Function to train and visualize SVM for all feature pairs and kernels
def svm_experiment(X, y, feature_pairs, kernels):
    for kernel in kernels:
        print(f'\n===== Using Kernel: {kernel} =====\n')
        for pair in feature_pairs:
            X_pair = X[:, pair]

            # Split data into train and test sets
            X_train, X_test, y_train, y_test = train_test_split(X_pair, y, test_size=0.3, random_state=0)

            if kernel == 'precomputed':
                # Create a Gram matrix (dot product)
                X_train_gram = np.dot(X_train, X_train.T)
                X_test_gram = np.dot(X_test, X_train.T)
                clf = SVC(kernel='precomputed')
                clf.fit(X_train_gram, y_train)
                y_pred = clf.predict(X_test_gram)
            else:
                # Train SVM classifier
                clf = SVC(kernel=kernel)
                clf.fit(X_train, y_train)
                y_pred = clf.predict(X_test)

            # Calculate accuracy and display confusion matrix
            accuracy = accuracy_score(y_test, y_pred)
            print(f'Feature Pair: {pair} | Accuracy: {accuracy:.3f}')

            cm = confusion_matrix(y_test, y_pred)
            disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
            disp.plot(cmap='viridis', values_format='.0f')
            plt.title(f'Confusion Matrix (Kernel: {kernel}) for Features: {pair}')
            plt.show()

            # Plot decision boundary only for non-precomputed kernels
            if kernel != 'precomputed':
                plot_decision_boundary(X_pair, y, pair, clf, kernel)

# Run the experiment
svm_experiment(X, y, feature_pairs, kernels)
```

From this we can say that the **best performing feature pair is (2,3)**

Features 2 (petal length) and 3 (petal width) are the most discriminative features for distinguishing between Iris species. This is well-known from data analysis of the Iris dataset.
The high accuracy for linear, poly, rbf, and precomputed kernels indicates that the classes are well-separated in 2D space for this pair.
The poor performance of sigmoid (11%) indicates that this kernel is not suitable for this combination, possibly because it behaves like a neural network activation and does not fit well to the simple boundary required here.

Feature Pairs (1, 3), (0, 3), (0, 2), (1, 2):     These features are somewhat effective but less discriminative than (2, 3).These features have moderate separability, which is why linear, rbf, and poly kernels perform well.
Sigmoid (24%) struggles here because it is more sensitive to data scaling and separability. It works well in high-dimensional spaces, not in simple 2D projections like these.

Worst Performing Feature Pair: (0, 1) : Features 0 (sepal length) and 1 (sepal width) do not provide clear separation of the classes.
The overlap of class clusters in the 2D space means that it is difficult for any classifier to create a proper decision boundary.
Even rbf and poly, which are non-linear kernels, cannot completely separate these clusters.
This is expected for the Iris dataset, as sepal length and sepal width are known to have lower discriminative power than petal features.

The RBF and linear kernels are the most robust, providing high accuracy across most feature pairs.
